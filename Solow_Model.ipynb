# Modelo de Crecimiento de Solow

Modelo de crecimiento de Solow, también conocido como modelo de crecimiento neoclásico, es un modelo macroeconómico creado para explicar el crecimiento económico y las variables que inciden en este en el largo plazo. El modelo se remonta a los trabajos de <a href="https://es.wikipedia.org/wiki/Robert_Solow">Robert Solow</a> y <a href="https://en.wikipedia.org/wiki/Trevor_Swan">Trevor Swan</a> en 1956. 

Dinámica de un modelo no lineal en diferencias

$$k_{t+1} = \psi(k_t)$$

donde,

* $\psi(k_t)= (1- \delta) k + s f(k)$

* $f(k) = k^{\alpha}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solow(α, s, δ, T, μ, σ):
    
    k0 = 1

    k = np.zeros(T)
    k[0] = k0
    ψ = k

    for i in range(T-1):
        shock = np.random.lognormal(μ,σ)
        y = k[i]**α
        k[i+1] = (1 - δ)*k[i] + shock*s*y
    
    ss = (s/δ)**(1/(1-α)) #Estado estacionario

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111)

    ax.plot(ψ, '-', color='red', label='Capital (k)')
    ax.plot(np.linspace(ss,ss,T), '--', color = 'black', label = 'Estado estacionario')


    ax.set_xlabel('tiempo')
    ax.set_ylabel('K')
    ax.legend()

    plt.show()
```

## Modelo Solow _sin shock_

```python
solow(0.3,0.2,0.1,80, False, False)
```

## Modelo Solow _con shock_

_shock_ con $\mu = 0$ y $\sigma = 0.2$

Distribución del _shock_ es lognormal donde:
* $\mu = e^{\mu + \sigma^2/2}$
* $\sigma^2 = (e^{\sigma^2} - 1)e^{\sigma^2\mu + \sigma^2}$

```python
solow(0.3,0.2,0.1,80, 0, 0.2)
```
